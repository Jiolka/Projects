Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

## Цель работы

Цель работы - провести исследовательский анализ полученных данных и решить следующие задачи:

 - Изучить следующие параметры: площадь, цена, число комнат, высота потолков, время продажи квартиры.

 - Выбрать 10 населённых пунктов с наибольшим числом объявлений, посчитать среднюю цену квадратного метра в этих населённых пунктах, выделить среди них населённые пункты с самой высокой и низкой стоимостью жилья.

## Рассчёты и добавление результатов в таблицу

Добавим в таблицу следующие столбцы:
- цена квадратного метра  - 'price_m'
- день недели - 'weekday' (с номером дня недели) и 'weekday_wr' (с названием)
- месяц - 'month'(с номером месяца) и 'month_wr' (с названием)
- год - 'year'
- этаж (первый, последний, другой) - 'floor_voc' и 'floor_gr'
- соотношение жилой и общей площади - 'living_vs_all'
- отношение площади кухни к общей - 'kitchen_vs_all'

```python
data['price_m'] = (((data['last_price'] / data['total_area']) * 100).astype('int')) / 100
```

```python
#Добавим день недели, месяц и год публикации
data['weekday'] = data['first_day_exposition'].dt.weekday
def weekday_wr(weekday):
    if weekday == 0:
        return 'Понедельник'
    if weekday == 1:
        return 'Вторник'
    if weekday == 2:
        return 'Среда'
    if weekday == 3:
        return 'Четверг'
    if weekday == 4:
        return 'Пятница'
    if weekday == 5:
        return 'Суббота'
    return 'Воскресенье'
    
data['weekday_wr'] = data['weekday'].apply(weekday_wr)

data['month'] = data['first_day_exposition'].dt.month
def month_wr(month):
    if month == 1:
        return 'Январь'
    if month == 2:
        return 'Февраль'
    if month == 3:
        return 'Март'
    if month == 4:
        return 'Апрель'
    if month == 5:
        return 'Май'
    if month == 6:
        return 'Июнь'
    if month == 7:
        return 'Июль'
    if month == 8:
        return 'Август'
    if month == 9:
        return 'Сентябрь'
    if month == 10:
        return 'Октябрь'
    if month == 11:
        return 'Ноябрь'
    return 'Декабрь'
data['month_wr'] = data['month'].apply(month_wr)

data['year'] = data['first_day_exposition'].dt.year
data.head()
```

```python
data['floors_total'].unique()
```

Добавим категории по этажам.

```python

def floor_voc(data):
    floor = data['floor']
    floor_total = data['floors_total']
    if floor == 1:
        return 1
    if floor == floor_total:
        return 3
    return 2 
 
data['floor_voc'] = data.apply(floor_voc, axis=1)

def floor_gr(data):
    floor = data['floor']
    floor_total = data['floors_total']
    if floor == 1:
        return 'Первый этаж'
    if floor == floor_total:
        return 'Последний этаж'
    return 'Другой этаж' 
 
data['floor_gr'] = data.apply(floor_gr, axis=1)
data.head()
```
